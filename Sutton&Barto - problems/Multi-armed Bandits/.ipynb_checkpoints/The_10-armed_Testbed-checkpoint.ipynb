# The 10 - armed Testbed

This is a similar excersice as that presented in section 2.3 of the K-Armed Bandit chapter, which makes an assessment of the (relative) effectiveness of the $\text{greedy}$ and the $\epsilon\text{-greedy}$ action value methods.

From section 2.3 of the book:

"To roughly assess the relative effectiveness of the greedy and $\epsilon\text{-greedy}$ action-value methods, we compared them numerically on a suite of test problems. This was a set of 2000 rendomly generated <i>k</i>-armed bandit problems with <i>k</i> = 10. For each bandit problem, such as the one shown in Figure 2.1, the action values, $q_*(a), a = 1, \ldots, 10$, were selected according to a normal (Gaussian) distribution with mean 0 and variance 1. Then, when a learning method applied to that problem selected action $A_t$ at time step $t$, the actual reward, $R_t$, was selected from a normal distribution with mean $q_*(A_t)$ and variance 1. These distributions are shown in gray in Figure 2.1. We call this suite of test and behavior as it improves with experience over 1000 time steps when applied to one of the bandit problems. This makes up one <i>run</i>. Repeating this for 2000 independent runs, each with a different bandit problem, we obtained measures of the learning algorithm's average behaviour."

I changed a bit the excersice. Instead of 1000 time steps, I am taking 10000 while I performing 2000 runs as described in the boo.

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from scipy.stats import norm
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (10, 6)

%load_ext autoreload
%autoreload 2

np.random.seed(1981)  # we can reproduce everything once and again
```

 ### Create the k-armed bandit

```python
### this class creates a k-armed bandit ###
class K_armed_bandit(object):
    def __init__(self, num_arms=10, mean=0, stdev=1):
        self.num_arms = num_arms
        self.actions = np.arange(num_arms)
        
        # q* for each arm
        self.q_substar = np.random.normal(loc=mean, scale=stdev, size=num_arms)
        self.optimal_action = np.argmax(self.q_substar)
        
        #potential feature, different variance for each arm
        self.stdev = stdev
        
    def get_num_arms(self):
        return self.num_arms
        
    def get_action_reward(self, action):
        return np.random.normal(self.q_substar[action], scale=self.stdev)
```

```python
# we actually create a k-armed bandit here as an example
bandit = K_armed_bandit()
```

```python
fig, axes = plt.subplots(1, bandit.get_num_arms(), sharex=True, sharey=True, gridspec_kw={'wspace':0, 'hspace':0})
axes = axes.ravel()

for k, q_m in enumerate(bandit.q_substar):
    ax = axes[k]
    #plot the distribution very close to the real one by sampling a big number of data
    ax.violinplot(np.random.normal(q_m, scale=bandit.stdev, size=500), showmeans=True)
    ax.set_xticks([])
    ax.set_xlabel('$q^*({0})$'.format(k + 1))
    xlim = ax.get_xlim()
    ax.plot([0,2],[0,0], '--', lw=1, color='black')
    ax.set_xlim(xlim)
    if k == 0:
        ax.set_ylabel('Reward distribution')
```

### Define a learning agent

Remember that the estimation of the action-value method used in this excersice is the sample-averages method (according to section 2.2 of the Multi-armed Bandits chapter):
$$Q_t(a) \doteq \frac{\text{sum of rewards when }a\text{ taken prior to }t}{\text{number of times }a\text{ taken prior to }t} = \frac{\sum_{i=1}^{t-1}\mathbb{R}_i\cdotp\mathbb{1}_{A_i=a}}{\sum_{i=1}^{t-1}\mathbb{1}_{A_i=a}},$$

where $\mathbb{1}_\text{predicate}$ denotes the random variable that is 1 if <i>predicate</i> is true and 0 if it is not.

```python
class Agent(object):
    def __init__(self, env, epsilon=0):
        self.env = env
        self.q = np.repeat(0., env.actions.shape[0]) # estimaged q
        
        # total sum of rewards collected for each arm
        self.total_rewards = np.repeat(0., env.actions.shape[0])
        # total count of visits to each action (to each arm)
        self.count_taken_arms = np.repeat(0., env.actions.shape[0])
        
        self.epsilon = epsilon
        self.reward_history = []
        self.action_history = []
        
    # we make a first initialization of each arm's q value.
    # this is done by 'actionating' each arm.
    # rounds = the number of times each arm is goint go be actionated during the
    # initialization process
    def initialize_q(self, rounds = 1):
        for _ in range(rounds):
            for action in self.env.actions:
                reward = self.env.get_action_reward(action)
                self.total_rewards[action] += reward
                self.count_taken_arms[action] += 1
        
        self.update_q()
    
    def update_q(self):
        self.q = self.total_rewards / self.count_taken_arms
        
    def step(self):
        action = self.select_action()
        
        reward = self.env.get_action_reward(action)
        self.total_rewards[action] += reward
        self.count_taken_arms[action] += 1
        self.q[action] = self.total_rewards[action] / self.count_taken_arms[action]
        self.reward_history.append(reward)
        self.action_history.append(action)
        
    def select_action(self):
        # chooses an action based on the epsilon value.
        # depending on the epsilon value the action will be choosen (and returned)
        # by following a greedy policy or an epsilon-greedy policy
        
        if np.random.random() >= self.epsilon:
            return np.argmax(self.q)
        else:
            return np.random.choice(self.env.actions)
        
    def expected_reward_with_experience(self):
        """calculate value_history based on reward_history, undiscounted"""
        return np.cumsum(self.reward_history) / (np.arange(len(self.reward_history)) + 1)
    
    def optimal_action_history(self):
        optimal_action = np.argmax(self.env.q_substar)
        act_hist = np.array(self.action_history)
        frac_hist = []
        for i in np.arange(act_hist.shape[0]  + 1):
            sub_hist = act_hist[:i]
            mask = sub_hist == optimal_action
            frac = sub_hist[mask].shape[0] / i
            frac_hist.append(frac)
        return frac_hist
```

### Assessment of $\text{greedy}$ vs $\epsilon\text{-greedy}$ policies

We generate 2000 independent k-armed bandits, each one of 10 arms.
The agent performs 1000 steps on each of those 2000 10-armed bandits.

```python
number_steps = 10000
number_runs = 2000

ave_rewards_dd = {}
frac_optimal_actions_dd = {}

epsilons = [0., 0.01, 0.04, 0.06, 0.1]

for eps in epsilons:
    expctd_rewards_list = [] # records value history of all steps
    frac_hist_list = []
    
    for _ in range(number_runs):
        agent = Agent(env=K_armed_bandit(), epsilon=eps)
        agent.initialize_q()
        for i in range(number_steps):
            agent.step() # perform 'number_steps' number of steps
        
        expctd_rewards_list.append(agent.expected_reward_with_experience())
        frac_hist_list.append(agent.optimal_action_history())
        
    expctd_rewards_array = np.array(expctd_rewards_list)
    ave_rewards_dd[eps] = expctd_rewards_array.mean(axis = 0)
    
    frac_hist_array = np.array(frac_hist_list)
    frac_optimal_actions_dd[eps] = frac_hist_array.mean(axis = 0)
```

```python
plt.figure(figsize=(10, 6))
colors = ['blue', 'red', 'green', 'orange', 'yellow']
for k, eps in enumerate(epsilons):
    col = colors[k]
    steps = np.arange(len(ave_rewards_dd[eps]))
    plt.plot(steps, ave_rewards_dd[eps], color=col, lw=1.5, label='epsilon = {0}'.format(eps))
plt.legend(loc='best')
plt.grid()
plt.xlabel('Number of Steps')
plt.ylabel('Average reward over past steps')
```

### Zomming in a bit

```python
plt.figure(figsize=(10, 6))
colors = ['blue', 'red', 'green', 'orange', 'yellow']
for k, eps in enumerate(epsilons):
    col = colors[k]
    steps = np.arange(len(ave_rewards_dd[eps]))
    plt.plot(steps, ave_rewards_dd[eps], color=col, lw=1.5, label='epsilon = {0}'.format(eps))
plt.grid()
plt.ylim(1.25, 1.5)
plt.xlim(-50, 5000)
plt.xlabel('Number of steps')
plt.ylabel('Average reward over past steps')
plt.legend(loc='lower right')
```

```python
plt.figure(figsize=(10, 6))
colors = ['blue', 'red', 'green', 'orange', 'yellow']
for k, eps in enumerate(epsilons):
    col = colors[k]
    steps = np.arange(len(frac_optimal_actions_dd[eps]))
    plt.plot(steps, frac_optimal_actions_dd[eps], color=col, lw=1.5, label='epsilon = {0}'.format(eps))
plt.grid()
plt.xlabel('Number of steps')
plt.ylabel('% optimal actions')
plt.legend(loc='lower right')
```
