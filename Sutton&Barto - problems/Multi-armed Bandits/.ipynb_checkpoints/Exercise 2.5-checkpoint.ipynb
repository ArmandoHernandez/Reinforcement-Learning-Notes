Design and conduct an experiment to demonstrate the difficulties that sample-average methods have for nonstationary problems. Use a modified vesion of the 10-armed testbed in which all the $q_*(a)$ start out equal and then take independent random walks (say by adding a normally distributed increment with mean zero and standard deviation 0.01 to all the $q_*(a)$ on each step). Prepare plots like Figure 2.2 (i.e. the plots on notebook [The 10-armed Testbed](./The_10-armed_Testbed.ipynb)) for an action-value method using sample averages, incrementally computed, and another action-value method using using a constant step-size parameter, $\alpha = 0.1$. Use $\epsilon = 0.1$ and longer runs, say  of 10,000 steps.

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
#from scipy.stats import norm
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (10, 6)

%load_ext autoreload
%autoreload 2

np.random.seed(1981)  # we can reproduce everything once and again
```

 ### Create the k-armed bandit

```python
### this class creates a k-armed bandit ###
class K_armed_bandit(object):
    def __init__(self, num_arms=10, stdev=1, init_val_qs=1.0):
        self.num_arms = num_arms
        self.actions = np.arange(num_arms)
        
        # q* for each arm
        self.q_substar = np.repeat(init_val_qs, num_arms)
        #self.optimal_action = np.argmax(self.q_substar)
        
        #potential feature, different variance for each arm
        self.stdev = stdev
        
    def get_num_arms(self):
        return self.num_arms
        
    def get_action_reward(self, action):
        # at each step, the agent wil receive a Reward from the environment (i.e. from the bandit)
        # The reward will be provided only once per step, so it is safe to update
        # the q_subsar values according to the requirment of this exercise.
        
        # first we calculate the reward we're going to return, before actually updating
        # the values of q substar
        rewards_to_return = np.random.normal(self.q_substar[action], scale=self.stdev)
        # calculate the independently increment that will be added to the true q values
        # (i.e. q_substar)
        distributed_increment = np.random.normal(loc=0, scale=0.01, size=self.num_arms)
        # add the increment calculated to q substar
        self.q_substar += distributed_increment
        # return the rewards
        return rewards_to_return
```

```python
# we actually create a k-armed bandit here as an example
bandit = K_armed_bandit()
```

```python
fig, axes = plt.subplots(1, bandit.get_num_arms(), sharex=True, sharey=True, gridspec_kw={'wspace':0, 'hspace':0})
axes = axes.ravel()

for k, q_m in enumerate(bandit.q_substar):
    ax = axes[k]
    #plot the distribution very close to the real one by sampling a big number of data
    ax.violinplot(np.random.normal(q_m, scale=bandit.stdev, size=500), showmeans=True)
    ax.set_xticks([])
    ax.set_xlabel('$q^*({0})$'.format(k + 1))
    xlim = ax.get_xlim()
    ax.plot([0,2],[0,0], '--', lw=1, color='black')
    ax.set_xlim(xlim)
    if k == 0:
        ax.set_ylabel('Reward distribution')
```

### Design an agent that make the following q value estimations in each step: sample average, incrementally computed and constant step-size parameter

```python
class Agent(object):
    def __init__(self, env, epsilon=0.1, alpha=0.1):
        self.env = env
        self.q = np.repeat(0., env.actions.shape[0]) # estimaged q
        
        # total sum of rewards collected for each arm
        self.total_rewards = np.repeat(0., env.actions.shape[0])
        # total count of visits to each action (to each arm)
        self.count_taken_arms = np.repeat(0., env.actions.shape[0])
        
        self.epsilon = epsilon
        self.alpha = alpha
        self.reward_history = []
        self.action_history = []
        
    # we make a first initialization of each arm's q value.
    # this is done by 'actionating' each arm.
    # rounds = the number of times each arm is goint go be actionated during the
    # initialization process
    def initialize_q(self, rounds = 1):
        for _ in range(rounds):
            for action in self.env.actions:
                reward = self.env.get_action_reward(action)
                self.total_rewards[action] += reward
                self.count_taken_arms[action] += 1
        
        self.update_q()
    
    def update_q(self):
        self.q = self.total_rewards / self.count_taken_arms
        
    def step_with_estimation_sample_average(self):
        action = self.select_action()
        
        reward = self.env.get_action_reward(action)
        self.total_rewards[action] += reward
        self.count_taken_arms[action] += 1
        self.q[action] = self.total_rewards[action] / self.count_taken_arms[action]
        self.reward_history.append(reward)
        self.action_history.append(action)
        
    def step_with_estimation_inc_computing(self):
        action = self.select_action()
        
        reward = self.env.get_action_reward(action)
        self.count_taken_arms[action] += 1
        self.q[action] += (reward - self.q[action]) / self.count_taken_arms[action]
        self.reward_history.append(reward)
        self.action_history.append(action)
        
    def step_with_constant_alpha(self):
        action = self.select_action()
        
        reward = self.env.get_action_reward(action)
        self.q[action] += (self.alpha * (reward - self.q[action]))
        self.reward_history.append(reward)
        self.action_history.append(action)
        
    def select_action(self):
        # chooses an action based on the epsilon value.
        # depending on the epsilon value the action will be choosen (and returned)
        # by following a greedy policy or an epsilon-greedy policy
        
        if np.random.random() >= self.epsilon:
            return np.argmax(self.q)
        else:
            return np.random.choice(self.env.actions)
        
    def expected_reward_with_experience(self):
        """calculate value_history based on reward_history, undiscounted"""
        return np.cumsum(self.reward_history) / (np.arange(len(self.reward_history)) + 1)
    
    def optimal_action_history(self):
        optimal_action = np.argmax(self.env.q_substar)
        act_hist = np.array(self.action_history)
        frac_hist = []
        for i in np.arange(act_hist.shape[0]  + 1):
            sub_hist = act_hist[:i]
            mask = sub_hist == optimal_action
            frac = sub_hist[mask].shape[0] / i
            frac_hist.append(frac)
        return frac_hist
```

### We run 2000 runs of 10000 steps each in three independen bandits (all 10-armed and initialized in 0 for all actions).
The runs will perform action value estimations:
    * one will perform sample-average method
    * one will perform incrementally computed method
    * one will perform a constan-alpha method

#### First we run a bandit with all action values initialized at 1.0

```python
number_steps = 10000
number_runs = 2000
num_estimation_methods = 3
methods = ['sampleAverage', 'incrementalComp', 'constantAlpha']

ave_rewards_1stExperiment = {}
frac_optimal_actions_1stExperiment = {}
ave_rewards_2stExperiment = {}
frac_optimal_actions_2stExperiment = {}

aveRewardsArray = [ave_rewards_1stExperiment, ave_rewards_2stExperiment]
fracOptimalActionsArray = [frac_optimal_actions_1stExperiment, frac_optimal_actions_2stExperiment]

for experimentNumber, _ in enumerate(aveRewardsArray):
    for indx_method, estimation_method in enumerate(methods):
        expctd_rewards_list = [] # records value history of all steps
        frac_hist_list = []

        print("Estimation method: {0}".format(estimation_method) )
        for _ in range(number_runs):
            agent = Agent(env=K_armed_bandit(init_val_qs=(0.0 + experimentNumber)))
            for step in range(number_steps):
                if indx_method == 0:
                    agent.step_with_estimation_sample_average()
                elif indx_method == 1:
                    agent.step_with_estimation_inc_computing()
                elif indx_method == 2:
                    agent.step_with_constant_alpha()

                #if step % 1000 == 0:
                   # print("    {0} steps performed.".format(step))

            expctd_rewards_list.append(agent.expected_reward_with_experience())
            frac_hist_list.append(agent.optimal_action_history())



        expctd_rewards_array = np.array(expctd_rewards_list)
        aveRewardsArray[experimentNumber][estimation_method] = expctd_rewards_array.mean(axis = 0)

        frac_hist_array = np.array(frac_hist_list)
        fracOptimalActionsArray[experimentNumber][estimation_method] = frac_hist_array.mean(axis = 0)
```

```python
fig = plt.figure(figsize=(10, 12))
colors = ['blue', 'red', 'green']

for plot in range(2):
    axSubplot = fig.add_subplot(2,1,plot + 1)
    axSubplot.title.set_text('Experiment {0}: q* of Bandit initialized to {0}'.format(plot, plot))
    for k, estimation_method in enumerate(methods):
        col = colors[k]
        steps = np.arange(len(aveRewardsArray[plot][estimation_method]))
        plt.plot(steps, aveRewardsArray[plot][estimation_method], color=col, lw=1.5,
                 label='epsilon = {0}'.format(estimation_method))
    plt.grid()
    plt.ylim(-0.1, 1.9)
    plt.xlim(-50, 10000)
    plt.xlabel('Number of steps')
    plt.ylabel('Average reward over past steps')
    plt.legend(loc='lower right')
```
